```python
tag_collection0 = db.execute('SELECT tag FROM music_message order by collection DESC  limit 100') ##收藏量前100的tag
tag_play0 = db.execute('SELECT tag FROM music_message order by play DESC  limit 100')    ###播放前100的tag
tag_comments0 = db.execute('SELECT tag FROM music_message order by comments DESC  limit 100') 

tag_collection = tag_collection0.fetchall()
tag_play= tag_play0.fetchall()
tag_comments= tag_comments0.fetchall()

my_dict = {'order_byCollect':tag_collection, 'order_byPlay':tag_play, 'order_byComment':tag_comments}
Tags_stack = pd.DataFrame(my_dict)
Tags_stack       ###
# res.fetchall()
# pd.DataFrame(res.fetchall())
```

```python
######处理tag数据，用于stack图的

import squarify
import pandas as pd
import matplotlib.pyplot as plt


###处理collect里的tag标签信息
tags_collect = []
dom_collect = []
for i in Tags_stack["order_byCollect"]:
#     print(i[0])
    c = i[0].split('-')
    for j in c:
        if j not in tags_collect:
            tags_collect.append(j)
        else:
            continue
for item in tags_collect:
    num = 0
    for i in Tags_stack["order_byCollect"]:
        type2 = i[0].split('-')
        for j in range(len(type2)):
            if type2[j] == item:
                num += 1
            else:
                continue
    dom_collect.append(num)
# 数据创建
data_collect= {'tags': tags_collect, 'num': dom_collect}
tag_collect_dict = {}
for i in range(len(tags_collect)):
    tag_collect_dict[tags_collect[i]] = dom_collect[i]  
print(tag_collect_dict)



###处理play里的tag标签信息
tags_collect = []
dom_collect = []
for i in Tags_stack["order_byPlay"]:
#     print(i[0])
    c = i[0].split('-')
    for j in c:
        if j not in tags_collect:
            tags_collect.append(j)
        else:
            continue
for item in tags_collect:
    num = 0
    for i in Tags_stack["order_byPlay"]:
        type2 = i[0].split('-')
        for j in range(len(type2)):
            if type2[j] == item:
                num += 1
            else:
                continue
    dom_collect.append(num)
# 数据创建

tag_play_dict = {}
for i in range(len(tags_collect)):
    tag_play_dict[tags_collect[i]] = dom_collect[i]  
print(tag_play_dict)



###处理Comment里的tag标签信息
tags_collect = []
dom_collect = []
for i in Tags_stack["order_byComment"]:
#     print(i[0])
    c = i[0].split('-')
    for j in c:
        if j not in tags_collect:
            tags_collect.append(j)
        else:
            continue
for item in tags_collect:
    num = 0
    for i in Tags_stack["order_byComment"]:
        type2 = i[0].split('-')
        for j in range(len(type2)):
            if type2[j] == item:
                num += 1
            else:
                continue
    dom_collect.append(num)
# 数据创建

tag_comment_dict = {}
for i in range(len(tags_collect)):
    tag_comment_dict[tags_collect[i]] = dom_collect[i]  
print(tag_comment_dict.values())
print(tag_collect_dict.values())
print(tag_play_dict.values())

style_list = ["欧美","治愈","浪漫","民谣","说唱","伤感","治愈","怀旧"]

style_set = set(style_list)

compare_style = ['Collect','Play','Comment']
style_dict = {'Collect':tag_collect_dict,'Play':tag_play_dict,'Comment':tag_comment_dict}
data_style_compare = pd.DataFrame(index= compare_style)


# frame = pd.DataFrame(data)
# df1 = frame.sort_values(by='num', ascending=False)[0:10]
```
